```python
import pandas as pd
import re
import matplotlib.pyplot as plt

simpleclean = pd.read_csv("origin_simpleclean.csv", names=["origin", "cleaned"], header=None)
simpleclean
```

```python
import pandas as pd
import re
import matplotlib.pyplot as plt

spookyhash = pd.read_csv("hash128bit.csv", names=["cleaned", "hashed128", "hashed64", "hashed32", "murmur"], header=None)
spookyhash = spookyhash[1:].reset_index().drop(['murmur', "index", "cleaned"], axis=1)
# spookyhash = spookyhash.drop(['murmur'], axis=1)
spookyhash
```

```python
spookyhash["origin"] = simpleclean.origin
spookyhash["origin size"] = spookyhash.origin.apply(lambda x : len(x))
spookyhash["hashed128 spooky size"] = spookyhash.hashed128.apply(lambda x : len(x))
spookyhash["hashed64 spooky size"] = spookyhash.hashed64.apply(lambda x : len(x))
spookyhash["hashed32 spooky size"] = spookyhash.hashed32.apply(lambda x : len(x))
spookyhash
```

```python
import matplotlib 
import matplotlib.pyplot as plt
font = {'size' : 13}

plt.rc('font', **font)

plt.figure(figsize=(12,7))
plt.xlabel('URLs/Values character size')
plt.title("Orginal URL vs. Spooky Hash Value")
plt.ylabel('Number of URLs/Values in log base')
plt.xlim(right = 300)
# plt.xscale("log")
plt.yscale("log")
plt.bar(list(spookyhash.groupby("origin size").count().index), list(spookyhash.groupby("origin size")["origin"].count()), alpha = 0.7)
plt.bar(list(spookyhash.groupby("hashed32 spooky size").count().index), list(spookyhash.groupby("hashed32 spooky size")["origin"].count()), alpha = 0.7)
plt.bar(list(spookyhash.groupby("hashed64 spooky size").count().index), list(spookyhash.groupby("hashed64 spooky size")["origin"].count()), alpha = 0.7)
plt.bar(list(spookyhash.groupby("hashed128 spooky size").count().index), list(spookyhash.groupby("hashed128 spooky size")["origin"].count()), alpha = 0.7)
plt.legend(('Original URL', 'Spooky Hash 32bit Value', 'Spooky Hash 64bit Value', 'Spooky Hash 128bit Value'))
plt.savefig('original_vs_spookyhashall.png')
```
